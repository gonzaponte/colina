```python
import numpy             as np
import pandas            as pd
import tables            as tb
import scipy.stats       as st
import matplotlib.pyplot as plt

from nbtools import auto_plot_style
from jupyfun import progressbar

in_range = lambda x, a, b: (x>=a) & (x<b)
```

```python
#%matplotlib inline
auto_plot_style()
```

```python
data = pd.read_csv("toy-mc/demo.csv", sep=" ", header=0)
extent = (-6.75*5, 6.75*5)*2
```

```python
for evt, df in data.groupby("event"):
    img = df.filter(regex="img_*").values.reshape(10,10)
    plt.figure()
    plt.imshow(img, origin="lower", extent=extent)
    plt.title(f"{df.x0.iloc[0]} {df.y0.iloc[0]}")
    if evt==10: break
```

```python
data2 = pd.read_csv("toy-mc/demo2.csv", sep=" ", header=None).values
extent = (-6.75*5, 6.75*5)*2
```

```python
for evt in data2:
    img = evt[-800**2:].reshape(800,800)
    plt.figure()
    plt.imshow(img, origin="lower", extent=extent)
    break
```

```python
for i in range(10):
    plt.figure(figsize=(15, 6))
    
    img = data.iloc[i:i+1].filter(regex="img_*").values.reshape(10,10)
    x, y = data.iloc[i, 1:3]
    plt.figure()
    plt.subplot(1, 2, 1)
    plt.imshow(img, origin="lower", extent=extent)
    plt.title(f"{x:.1f} {y:.1f}")

    plt.subplot(1, 2, 2)
    img = data2[i, -80**2:].reshape(80,80)
    plt.imshow(img, origin="lower", extent=extent)
    plt.tight_layout()
```
